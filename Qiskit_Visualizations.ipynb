```python
from qiskit import *
from qiskit.visualization import plot_histogram
```

```python
# As an example we make a 2-qubit Bell state
```

```python
bell = QuantumCircuit(2, 2)
bell.h(0)
bell.cx(0, 1)
meas = QuantumCircuit(2, 2)
meas.measure([0,1], [0,1])

backend = BasicAer.get_backend('qasm_simulator')
circ = bell.compose(meas)
result = backend.run(transpile(circ, backend), shots=1000).result()
counts = result.get_counts(circ)
print(counts)
```

```python
plot_histogram(counts)
```

```python
second_result = backend.run(transpile(circ, backend), shots=1000).result()
second_counts = second_result.get_counts(circ)
third_result = backend.run(transpile(circ, backend), shots=1000).result()
third_counts = third_result.get_counts(circ)
legend = ['First execution', 'Second execution', 'third counts']
plot_histogram([counts, second_counts, third_counts], legend=legend)
```

```python
plot_histogram([counts, second_counts, third_counts], legend=legend, sort='desc', figsize=(15, 12), color=['orange', 'black', 'red'], bar_labels=False)
```

```python
from qiskit.visualization import plot_state_city, plot_bloch_multivector, plot_state_paulivec, plot_state_hinton, plot_state_qsphere
```

```python
backend = BasicAer.get_backend('statevector_simulator')
result = backend.run(transpile(bell, backend)).result()
psi = result.get_statevector(bell)
plot_state_city(psi)
```

```python
plot_state_hinton(psi)
```
